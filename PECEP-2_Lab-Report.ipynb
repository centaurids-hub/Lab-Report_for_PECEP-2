**LAB REPORT FOR REFLECTION 1.2: NUMBER SYSTEM CONVERSION**

```python
# NUMBER 1

num = float(input("Enter decimal number: "))

i = int(num)
f = num - i

def frac_convert(f, base, n=6):
    s = ""
    for _ in range(n):
        f *= base
        d = int(f)
        s += hex(d)[2:].upper()
        f -= d
    return s

print("Entered Decimal:", num)
print("Binary:", bin(i)[2:] + "." + frac_convert(f, 2))
print("Octal:", oct(i)[2:] + "." + frac_convert(f, 8))
print("Hex:", hex(i)[2:].upper() + "." + frac_convert(f, 16))
```

```python
# NUMBER 2

h = input("Enter hex: ").upper()

if '.' in h:
    i, f = h.split('.')
else:
    i, f = h, ""

# to decimal
dec = int(i, 16) + sum(int(x, 16) * 16**-(n+1) for n, x in enumerate(f))

def to_base(x, base, n=6):
    i = int(x)
    f = x - i
    s = bin(i)[2:] if base == 2 else oct(i)[2:]
    frac = ""
    for _ in range(n):
        f *= base
        d = int(f)
        frac += str(d)
        f -= d
    return s + "." + frac
print("Entered Hex:", h)
print("Decimal:", dec)
print("Binary:", to_base(dec, 2))
print("Octal:", to_base(dec, 8))
```

**LAB REPORT FOR COMBINATION**

```python
# Combination formula: nCr = n! / (r!(n-r)!)

import math

n = int(input("Enter n: "))
r = int(input("Enter r: "))

result = math.factorial(n) // (math.factorial(r) * math.factorial(n - r))

print("Entered n =", n)
print("Entered r =", r)
print("Combination =", result)
```

**LAB REPORT FOR PERMUTATION**

```python
# Permutation formula: nPr = n! / (n-r)!

import math

n = int(input("Enter n: "))
r = int(input("Enter r: "))

result = math.factorial(n) // math.factorial(n - r)

print("Entered n =", n)
print("Entered r =", r)
print("Permutation =", result)
```

**LAB REPORT FOR COIN PROBLEM**

```python
# Probability of getting a certain number of heads

import math

coins = int(input("Number of coins: "))
heads = int(input("Desired number of heads: "))

# Combination formula
ways = math.factorial(coins) // (
    math.factorial(heads) * math.factorial(coins - heads)
)

total_outcomes = 2 ** coins

probability = ways / total_outcomes

print("Number of coins:", coins)
print("Desired number of heads:", heads)
print("Probability =", probability)
```

**LAB REPORT FOR CARD PROBLEM**

```python
# Probability of getting a desired card
# With Replacement and Without Replacement

desired = int(input("Number of desired cards: "))
total_cards = 52

draws = int(input("Number of draws: "))

# WITH replacement
with_replacement = (desired / total_cards) ** draws

# WITHOUT replacement
without_replacement = 1

for i in range(draws):
    without_replacement *= (desired - i) / (total_cards - i)

print("Number of desired cards:", desired)
print("Number of draws:", draws)
print("Probability with replacement =", with_replacement)
print("Probability without replacement =", without_replacement)
```

**LAB REPORT FOR DICE**

```python
# Probability of getting a specific sum from rolling dice

from itertools import product

dice = int(input("Number of dice: "))
target_sum = int(input("Desired sum: "))

# All possible outcomes
outcomes = list(product(range(1, 7), repeat=dice))

# Count favorable outcomes
favorable = 0

for outcome in outcomes:
    if sum(outcome) == target_sum:
        favorable += 1

total_outcomes = len(outcomes)

probability = favorable / total_outcomes

print("Number of dice:", dice)
print("Desired sum:", target_sum)
print("Favorable outcomes =", favorable)
print("Total outcomes =", total_outcomes)
print("Probability =", probability)
```

**LAB REPORT FOR STATISTICS**

```python
# Mean, Median, Mode, Variance, and Standard Deviation
# User can keep entering numbers until typing "done"

import statistics

numbers = []

while True:
    data = input("Enter a number (type 'done' to finish): ")

    if data.lower() == "done":
        break

    numbers.append(float(data))

# Calculations
mean = statistics.mean(numbers)
median = statistics.median(numbers)
mode = statistics.mode(numbers)
variance = statistics.variance(numbers)
std = statistics.stdev(numbers)

# Output
print("\nData:", numbers)
print("Mean =", mean)
print("Median =", median)
print("Mode =", mode)
print("Variance =", variance)
print("Standard Deviation =", std)
```
